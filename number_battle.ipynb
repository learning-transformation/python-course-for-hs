```python
import random
import time

# 自分のHPと敵のHPを受け取ってその値を表示する関数
def showHP(myHP, enemyHP):
    print(f"{'あなたのHP':>12}({myHP:2}) ", end="")
    print('*' * myHP)
    print(f"{'敵のHP':>12}({enemyHP:2}) ", end="")
    print('+' * enemyHP)
    print("\n")

def battle(myAttackPoint, enemyAttackPoint, myHealthPoint, enemyHealthPoint, Yuragi, log):
    # あなたの攻撃
    hp = myAttackPoint - int(myAttackPoint * Yuragi * (random.randint(1, 99) / 100))
    enemyHealthPoint = max(0, enemyHealthPoint - hp)
    print(f"あなたの攻撃！{hp}のダメージを与えた！\n")

    # ログに記録
    log.append({
        "round": len(log) + 1,
        "action": "あなたの攻撃",
        "damage": hp,
        "myHP": myHealthPoint,
        "enemyHP": enemyHealthPoint
    })

    # 自分のHPと敵のHPを表示する関数を呼び出す
    showHP(myHealthPoint, enemyHealthPoint)

    if enemyHealthPoint <= 0:
        return myHealthPoint, enemyHealthPoint

    # 敵の攻撃
    hp = enemyAttackPoint - int(enemyAttackPoint * Yuragi * (random.randint(1, 99) / 100))
    myHealthPoint = max(0, myHealthPoint - hp)
    print(f"敵の攻撃！{hp}のダメージを受けた！\n")

    # ログに記録
    log.append({
        "round": len(log) + 1,
        "action": "敵の攻撃",
        "damage": hp,
        "myHP": myHealthPoint,
        "enemyHP": enemyHealthPoint
    })

    # 自分のHPと敵のHPを表示する関数を呼び出す
    showHP(myHealthPoint, enemyHealthPoint)

    return myHealthPoint, enemyHealthPoint

TotalPoint = 100
BattleTimes = 3
Yuragi = 0.5

# 自分の攻撃力とHPを決める
myAttackPoint = int(input("攻撃力を入れてください (1-99)\n"))

if myAttackPoint < 1 or myAttackPoint > 99:
    print("1-99の値を入れてください")
else:
    myHealthPoint = TotalPoint - myAttackPoint

    response = input(f"あなたの攻撃力: {myAttackPoint}, HP: {myHealthPoint}\n戦う?(Y/N)")

    if response.lower() != 'y':
        print("ゲーム終了")
    else:
        # 戦闘ログを記録する配列
        battle_log = []

        # 敵の攻撃力とHPをランダムに決める
        enemyAttackPoint = random.randint(1, TotalPoint - 1)
        enemyHealthPoint = TotalPoint - enemyAttackPoint
        print(f"敵の攻撃力: {enemyAttackPoint}, 敵HP: {enemyHealthPoint}\n\n")

        for i in range(BattleTimes):
            myHealthPoint, enemyHealthPoint = battle(myAttackPoint, enemyAttackPoint, myHealthPoint, enemyHealthPoint, Yuragi, battle_log)

            if enemyHealthPoint == 0 or myHealthPoint == 0:
                break

            time.sleep(2)

        if myHealthPoint >= enemyHealthPoint:
            print("あなたの勝ち！")
        else:
            print("あなたの負け！")

        # 戦闘ログを表示
        print("\n戦闘ログ:")
        for log in battle_log:
            print(f"ラウンド {log['round']} - {log['action']} - ダメージ: {log['damage']}, あなたのHP: {log['myHP']}, 敵のHP: {log['enemyHP']}")
```
